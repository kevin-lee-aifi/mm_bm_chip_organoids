```python
import sys
sys.path.insert(0, '/home/workspace/TissDiss')

from tissdiss_py_util import *

hdir = '/home/workspace/'
wdir = hdir + "/TissDiss/EXP-01244"
pltdir = wdir + "/EXP-01244_plots/"
srldir = wdir + "/EXP-01244_serial_obj/"

sub_adata = sc.read_h5ad(srldir + 'subclustered_adata.h5ad')
adata = sc.read_h5ad(srldir + 'processed_adata.h5ad')
adata = adata[adata.obs['harmony'] == 'no harmony']
```

# PAGA sample type plot

```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots(figsize=(15, 15))

# Create PAGA plot
sc.pl.paga(adata, color="sample_type", threshold=0.1, ax=ax, show=False, node_size_scale=12)

# Get unique sample types and their colors from the plot, but sort them
sample_types = sorted(adata.obs['sample_type'].unique())
colors = [adata.uns['sample_type_colors'][list(adata.obs['sample_type'].cat.categories).index(st)] for st in sample_types]

handles = []
labels = []
for sample_type, color in zip(sample_types, colors):
   handles.append(plt.scatter([], [], c=color, s=100))
   labels.append(sample_type)

ax.legend(handles, labels, loc='lower left')

fig.savefig(pltdir + 'sample_type_PAGA.pdf', bbox_inches='tight', dpi=300)
```

# PAGA sample type overlayed on UMAP

- The cluster label positions on the PAGA overlay is different than the labels on just the UMAP plot.
    - Due to disperse clusters; labels positioned in the middle of entire cluster.

```python
fig, ax = plt.subplots(figsize=(15, 15))

pos = pd.DataFrame(adata.obsm["X_umap"], index=adata.obs_names)
pos["group"] = adata.obs[adata.uns["paga"]["groups"]]
pos = pos.groupby("group", observed=True).mean()

sc.pl.umap(adata, show=False, ax=ax)

sample_types = sorted(adata.obs['sample_type'].unique())
colors = [adata.uns['sample_type_colors'][list(adata.obs['sample_type'].cat.categories).index(st)] for st in sample_types]

handles = []
labels = []
for sample_type, color in zip(sample_types, colors):
   handles.append(plt.scatter([], [], c=color, s=100))
   labels.append(sample_type)

ax.legend(handles, labels, loc='lower left')

sc.pl.paga(
    adata,
    color="sample_type",
    threshold=0.2,
    node_size_scale=4,
    edge_width_scale=0.7,
    pos=pos.values,
    random_state=0,
    ax=ax,
)

fig.savefig(pltdir + 'sample_type_PAGA_UMAP.pdf', bbox_inches='tight', dpi=300)
```
